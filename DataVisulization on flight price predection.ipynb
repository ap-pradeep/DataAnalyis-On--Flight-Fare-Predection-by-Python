```python
#import the libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import pandas as pd

# Assuming 'cleaned_data' is your DataFrame
df = pd.read_csv(r'C:\Users\anipi\Desktop\Summer Internship\Flight_Data (2).csv',encoding = 'unicode_escape', low_memory = False)
```

### Data Visulization

#### Flight Wise Analysis

```python
#plot data
data_airline_count = df.groupby(['airline'], as_index = False)['price'].count()
print("Airline wise price count \n", data_airline_count)

data_airline_duration = df.groupby(['airline'], as_index = False)['duration'].sum()
print("Airline was duration count \n" , data_airline_duration)
```

```python
#plotting
fig,ax = plt.subplots(1,2, figsize = (15,8))
#pie chart
ax[0].pie(x = data_airline_count['price'],
                    labels = data_airline_count['airline'], autopct = "%1.2f%%")
ax[0].set_title("airline wise price")
#Bar chart
sns.barplot(x = 'airline' ,y = 'price', data = data_airline_count)
ax[1].set_title("airline wise price")
plt.show()
```

### Conclusion

#### Vistara airline has the highest price and next comes Air_india

```python
#plotting
fig,ax = plt.subplots(1,2, figsize = (15,8))
#pie chart
ax[0].pie(x = data_airline_count['price'],
                    labels = data_airline_count['airline'], autopct = "%1.2f%%")
ax[0].set_title("airline wise price")
#Bar chart
sns.barplot(x = 'airline' ,y = 'duration', data = data_airline_duration)
ax[1].set_title("airline wise duration")
plt.show()
```

### Conclusion

#### Airline Vistara has highest time duration

### Source city Wise Analysis

```python
#plot data
data_airline_source_city = df.groupby(['source_city'], as_index = False)['price'].count()
data_airline_source_city
```

```python
# Plots
fig, ax = plt.subplots(1,2, figsize = (20,8))

# Countplot
sns.countplot( x = 'source_city', data = df, ax = ax[0])
ax[0].set_title('Source_city')

# Barplot for 
sns.barplot(y = 'price', x = 'source_city',
            data = data_airline_source_city, ax = ax[1])
ax[1].set_title('Price - Source_city')

plt.show()

```
